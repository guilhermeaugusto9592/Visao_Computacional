# Detecção de Rostos em uma imagem

Importar a biblioteca OpenCV

```python
import cv2
```

Carrega o classificador de detecção de rostos pré-treinado

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

Carrega a imagem

```python
img = cv2.imread('/content/drive/MyDrive/fotos/28287431-grupo-do-multietnico-o-negocio-pessoas-dentro-a-escritorio-ai-gerado-foto.jpg')
```

Detecta rostos na imagem

```python
faces = face_cascade.detectMultiScale(img, 1.3, 5)
```

Desenha retângulos ao redor dos rostos detectados

```python
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
```

Exibe a imagem com os rostos detectados

```python
import matplotlib.pyplot as plt

# Supondo que 'img' seja a imagem que você deseja exibir
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Desativa os eixos
plt.show()
```
